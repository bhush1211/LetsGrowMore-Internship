### Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

### Load image

```python
img=cv2.imread("balganesha.jpg")
```

```python
# Read the image in RGB format
img_rbg=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,6))
plt.imshow(img_rbg)
plt.title("Colored Image")
plt.axis("off")
```

```python
# Covert the original image to grayscale image
grayimage=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,6))
plt.imshow(grayimage, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
```

```python
# Inverting image
inv_img=255-grayimage
plt.figure(figsize=(10,6))
plt.imshow(inv_img,cmap='gray')
plt.title("Inverted Image")
plt.axis("off")
```

```python
# Blur image
blur=cv2.GaussianBlur(inv_img,(21,21),0)
plt.figure(figsize=(10,6))
plt.imshow(blur,cmap='gray')
plt.title('Smoothen Image')
plt.axis('off')
```

### Convert image to pencil sketch

```python
skh=cv2.divide(grayimage,255-blur,scale=255)
plt.figure(figsize=(10,6))
plt.imshow(skh,cmap='gray')
plt.title("Final Sketch")
plt.axis('off')
```
